```python
import pandas as pd
from faker import Faker
import random

fake = Faker()

# Define the number of players you want to generate
num_players = 250

# Define a list of nationalities
nationalities = ["Indian","Overseas"]

# Generate fake data for the Player table
players = []
team_id_counter = 1
players_per_team = 25

for player_id in range(1, num_players + 1):
    player_name = fake.name_male()
    base_price = random.randint(100000, 10000000)
    age = random.randint(18, 40)
    nationality = random.choice(nationalities)
    team_id = team_id_counter

    player = (player_id, player_name, base_price, age, nationality, team_id)
    players.append(player)

    if player_id % players_per_team == 0:
        team_id_counter += 1

# Create a DataFrame from the list of players
df = pd.DataFrame(players, columns=["player_id", "player_name", "base_price", "age", "nationality", "team_id"])

# Print the DataFrame
print(df)
```

```python
import pandas as pd
from faker import Faker
import random

fake = Faker()

# Define the number of players
num_players = 250

# Generate a list of unique player IDs
all_player_ids = list(range(1, num_players + 1))

# Shuffle the player IDs to distribute them randomly among the tables
random.shuffle(all_player_ids)

# Number of players to assign to each table
num_players_per_table = num_players // 3

# Batsman table
batsmen = []
for batsman_id in range(1, num_players_per_table + 1):
    player_id = all_player_ids.pop()
    strike_rate = round(random.uniform(50, 200), 2)
    no_of_fours = random.randint(0, 50)
    no_of_sixes = random.randint(0, 20)
    no_of_50s = random.randint(0, 10)
    no_of_100s = random.randint(0, 5)
    total_runs = random.randint(0, 5000)

    batsman = (player_id, batsman_id, strike_rate, no_of_fours, no_of_sixes, no_of_50s, no_of_100s, total_runs)
    batsmen.append(batsman)

# Bowler table
bowlers = []
for bowler_id in range(1, num_players_per_table + 1):
    player_id = all_player_ids.pop()
    no_of_wickets = random.randint(0, 500)
    type_of_bowler = random.choice(["Fast", "Spin", "Medium Pace"])

    bowler = (player_id, bowler_id, no_of_wickets, type_of_bowler)
    bowlers.append(bowler)

# Fielder table
fielders = []
for fielder_id in range(1, num_players_per_table + 1):
    player_id = all_player_ids.pop()
    no_of_stumping = random.randint(0, 20)
    no_of_catches = random.randint(0, 50)
    no_of_run_out = random.randint(0, 10)

    fielder = (player_id, fielder_id, no_of_stumping, no_of_catches, no_of_run_out)
    fielders.append(fielder)

# Create DataFrames from the lists of players
df_batsman = pd.DataFrame(batsmen, columns=["player_id", "batsman_id", "strike_rate", "no_of_fours", "no_of_sixes", "no_of_50s", "no_of_100s", "total_runs"])
df_bowler = pd.DataFrame(bowlers, columns=["player_id", "bowler_id", "no_of_wickets", "type_of_bowler"])
df_fielder = pd.DataFrame(fielders, columns=["player_id", "fielder_id", "no_of_stumping", "no_of_catches", "no_of_run_out"])

# Print the DataFrames
print("Batsman Table:")
print(df_batsman)

print("\nBowler Table:")
print(df_bowler)

print("\nFielder Table:")
print(df_fielder)
```
